```julia
using Plots
using FFTW
```

```julia
include("./SimplePSD.jl")
using .SimplePSD
```

```julia
carrierFreq  = 5.6007999999996275e9
samplingRate = 16e6
gain = 30.0
```

```julia
xAx, psd = SimplePSD.computePSD(carrierFreq,samplingRate,gain;args="",N=1024,nbMean=32)
```

```julia
plot(xAx,20*log10.(abs2.(fftshift(fft(psd)))))
```

```julia
plot(10*log10.(psd))
```

```julia
plot(xAx,10*log10.(abs2.(psd)))
```

```julia
plot(xAx,10*log10.(abs2.(fftshift(psd))))
```
